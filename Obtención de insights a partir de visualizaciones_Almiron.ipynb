En el caso del modelo 1 (regresión lineal) el valor de R-cuadrado = 0.57, obtenido indica que su bondad o aptitud no es perfecta (se encuentra considerablemente alejado de ser igual a 1). De todas formas podríamos pensar que el modelo es capaz de explicar alrededor del 57% de la variabilidad en el precio de venta de los inmuebles (variable dependiente).


Parece ser que, por amplia diferencia, el modelo 2 (decision tree) es el que mejor nos permite explicar a la variable dependiente (precio) en función del comportamiento de las variables independientes involucradas en el estudio. El R2 de 0.99 indica que el 99% de la variabilidad en el precio puede ser explicada por nuestro modelo.


Al mismo tiempo, las demás métricas calculadas (RMSE y MAE), resultan en la misma dirección dando mejores valores para el modelo 2 (considerablemente inferiores que en 1 y 3 ).


El satisfactorio valor de RMSE obtenido para el caso del decision tree nos habla de la bondad del ajuste, en un sentido absoluto. Esta métrica nos ayuda a asegurarnos que los resultados predichos no estan demasiado desviados con respecto al valor real. Algo similar ocurre con el valor de MAE calculado, que siendo tan cercano a 0 (cero), nos da un buen panorama en relación a la suma de error del modelo y pordríamos valernos de que la diferencia absoluta entre el valor objetivo y el valor predicho por nuestro modelo es pequeña.



El modelo de KNN regressor parecería tener dificultades a la hora de identificar patrones, y quizás eso sea la causa de los resultados poco alentadores.

A su vez, estas dos variables (cantidad de baños y habitaciones), parecen presentar una tendencia a incrementarse una cuando aumenta la otra.


En el Gráfico 4 se visualiza una relacion entre el numero de dormitorios y el numero de baños que parece tener una forma lineal, pudiendose trazar una recta por lo que seria la parte "mas central" de los datos. Hay una tendencia donde a medida que se incrementa un valor lo hace tambien el otro, por lo que los puntos muestran una relacion entre si.
